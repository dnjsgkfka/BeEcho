```python
import glob
import os
import random
from sklearn.model_selection import train_test_split

base_path = '/content/drive/MyDrive/be-echo-dataset'

# 'images' ë¶ˆëŸ¬ì˜¤ê¸°
image_path = os.path.join(base_path, 'images')
all_images = glob.glob(os.path.join(image_path, '*.JPG')) + \
             glob.glob(os.path.join(image_path, '*.jpg')) + \
             glob.glob(os.path.join(image_path, '*.jpeg')) + \
             glob.glob(os.path.join(image_path, '*.png'))

if len(all_images) == 0:
    print("error! file not found")

else:
    # train:val:test = 8:1:1
    train_images, temp_images = train_test_split(all_images, test_size=0.2, random_state=42)
    val_images, test_images = train_test_split(temp_images, test_size=0.5, random_state=42)

    with open('/content/train.txt', 'w') as f:
        f.write('\n'.join(train_images))

    with open('/content/val.txt', 'w') as f:
        f.write('\n'.join(val_images))

    with open('/content/test.txt', 'w') as f:
        f.write('\n'.join(test_images))

    print(f"\nì´ë¯¸ì§€ ì´ {len(all_images)}ì¥")
    print(f"train: {len(train_images)}ì¥, val: {len(val_images)}ì¥, test: {len(test_images)}")
```

```python
import yaml

# label ì´ë¦„
class_names = ['tumbler', 'disposable_cup']

data_yaml = {
    'train': 'train.txt',  # trainìš© íŒŒì¼
    'val': 'val.txt',      # validationìš© íŒŒì¼
    'test': 'test.txt',    # testìš© íŒŒì¼
    'nc': len(class_names),
    'names': class_names
}

# 'eco_data.yaml' íŒŒì¼ ìƒì„±
with open('eco_data.yaml', 'w') as f:
    yaml.dump(data_yaml, f)

print("eco_data.yaml file created")
```
